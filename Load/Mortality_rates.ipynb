```python
import os
import csv
import pandas as pd
import numpy as np
```

```python
#Collecting data from Resources folder

#Mortality rates

file_path = os.path.join("..","References", "Excess_mortality_during_COVID-19_pandemic.csv")

excess_mortality_df = pd.read_csv(file_path)

excess_mortality_df.head()
```

```python
excess_mortality_df.describe()
```

```python
#Finding unique locations
excess_mortality_df["country"].unique()
```

```python
#remove all countries except US
clean_df = excess_mortality_df[~excess_mortality_df["country"].str.contains('Italy')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Netherlands')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Ecuador')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Spain')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('France')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Turkey')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Indonesia')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Russia')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Brazil')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Peru')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Chile')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Mexico')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('UK')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Belgium')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Portugal')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Denmark')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Austria')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Germany')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Sweden')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('South Africa')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Norway')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Israel')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Iceland')]
clean_df = clean_df[~excess_mortality_df["country"].str.contains('Switzerland')]

clean_df["country"].unique()
```

```python
clean_df["year"].unique()
```

```python
clean_US_df = clean_df[~clean_df["year"].str.contains('2013')]
clean_US_df = clean_US_df[~clean_df["year"].str.contains('2014')]
clean_US_df = clean_US_df[~clean_df["year"].str.contains('2015')]
clean_US_df = clean_US_df[~clean_df["year"].str.contains('2016')]
clean_US_df = clean_US_df[~clean_df["year"].str.contains('2017')]
clean_US_df = clean_US_df[~clean_df["year"].str.contains('2018')]
clean_US_df = clean_US_df[~clean_df["year"].str.contains('2019')]

clean_US_df["year"].unique()
```

```python
clean_US_df.head(50)
```

```python
clean_US_final_df= clean_US_df.groupby(['month']).agg({"deaths":'max',
                                                   "expected_deaths":'max',
                                                       "excess_deaths": 'max',
                                                        })
clean_US_final_df.head()
```

```python
clean_US_final1_df = clean_US_final_df.drop(index = '#date+reported+mm')
clean_US_final1_df.head(10)
```

```python
clean_US_final2_df = clean_US_final1_df.rename(index = {'1': 'Jan 2020', '2': 'Feb 2020', '3': 'Mar 2020', '4': 'Apr 2020', '5': 'May 2020', '6': 'Jun 2020', '7': 'July 2020', '8': 'Aug 2020' })
clean_US_final2_df.head()
```

```python
clean_US_final2_df['deaths'] = clean_US_final2_df.deaths.astype(int)
clean_US_final2_df['expected_deaths'] = pd.to_numeric(clean_US_final2_df['expected_deaths'])
clean_US_final2_df['excess_deaths'] = pd.to_numeric(clean_US_final2_df['excess_deaths'])
```

```python
clean_US_final2_df.dtypes
```

```python
cols = ['expected_deaths', 'excess_deaths']
clean_US_final2_df[cols] = clean_US_final2_df[cols].applymap(np.int64)
print(clean_US_final2_df)
```
