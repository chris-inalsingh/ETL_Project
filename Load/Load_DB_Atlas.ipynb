```python
## Custom Class which confirms datatype is readable by MongoBD
# and if test fail, convert to readable variable


class CustomEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, numpy.integer):
            return int(obj)
        elif isinstance(obj, numpy.floating):
            return float(obj)
        elif isinstance(obj, numpy.ndarray):
            return obj.tolist()
        else:
            return super(CustomEncoder, self).default(obj)
```

```python
#Covid Infection Collection        
world_data_dict=world_data_4_df.to_dict("records")
world_clean_json = json.dumps(world_data_dict,cls=CustomEncoder)
world_final  = json.loads(world_clean_json)

#Country Collection
country_data_dict=Country_df.to_dict("records") ###
country_clean_json = json.dumps(country_data_dict,cls=CustomEncoder)
country_final  = json.loads(country_clean_json)

#Stringency Collection
string_data_dict=data_df.to_dict("records")
string_clean_json = json.dumps(string_data_dict,cls=CustomEncoder)
string_final  = json.loads(string_clean_json)

#Excess Death Collection
xsdeath_data_dict=US_by_region_df.to_dict("records")
xsdeath_clean_json = json.dumps(xsdeath_data_dict,cls=CustomEncoder)
xsdeathfinal  = json.loads(xsdeath_clean_json)

#Unemployment Rate Collection
unemp_data_dict=Unemployment_Rates.to_dict("records")
unemp_clean_json = json.dumps(unemp_data_dict,cls=CustomEncoder)
unemp_final  = json.loads(unemp_clean_json)



#Influenza Collection
flu_data_dict=Influenza_table.to_dict("records")
flu_clean_json = json.dumps(flu_data_dict,cls=CustomEncoder)
flu_final  = json.loads(flu_clean_json)
```
